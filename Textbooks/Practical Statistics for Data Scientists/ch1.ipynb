```python
import pandas as pd
import numpy as np
```

```python
stateDf = pd.read_csv('./data/state.csv')
```

## Single Variable Statistics

```python
stateDf.info()
```

```python
stateDf.describe()
```

```python
print("Murder Rate Statistics Among States")
print("Murder rate variance: {}".format(stateDf['Murder.Rate'].var()))
print("Murder rate std: {}".format(stateDf['Murder.Rate'].std()))
print("Murder rate mean absolute deviation: {}".format(stateDf['Murder.Rate'].mad()))
print("Murder rate mean absolute deviation: {}".format(stateDf['Murder.Rate'].mad()))
```

### Order Statistics

```python
stateMR = stateDf['Murder.Rate']
```

```python
stateMR = stateMR.sort_values()
```

```python
stateMR
```

```python
highest_murder_rate = stateDf[stateDf["Murder.Rate"] == stateDf["Murder.Rate"].max()]
print("State with the highest murder rate is {} with a murder rate of {}".format(highest_murder_rate['State'].values[0], highest_murder_rate['Murder.Rate'].values[0]))
```

```python
print("Median: {}".format(stateDf['Murder.Rate'].quantile(.50)))
print("Q1: {}".format(stateDf["Murder.Rate"].quantile(.25)))
print("Q3: {}".format(stateDf["Murder.Rate"].quantile(.75)))
print("IQR: {}".format(stateDf["Murder.Rate"].quantile(.75) - stateDf['Murder.Rate'].quantile(.25)))
```

## Single Variable Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.title('Murder Rate By State', fontsize='16', pad=10)
sns.barplot(data=stateDf.sort_values(by='Murder.Rate', ascending=False), x='State', y='Murder.Rate')
```

```python
sns.boxplot(stateDf['Murder.Rate'])
```

```python
sns.distplot(stateDf['Murder.Rate'],bins=20)
```
